```python
import pandas_datareader.data as web
import datetime
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(30,20))
import numpy as np
```

## Get Data for 'GOOG' and 'SPY'

```python
start = datetime.datetime(2017, 1, 1)
end = datetime.datetime(2017, 12, 21)
goog_data = web.DataReader('GOOG','yahoo',start, end)
spy_data = web.DataReader('SPY','yahoo',start, end)
```

```python
fig, axs = plt.subplots(1,2)
goog_data[['Adj Close']].plot(ax=axs[0],figsize=(15,5),title="Google")
spy_data[['Adj Close']].plot(ax=axs[1],figsize=(15,5),title="SnP")
```

## Calculate Daily Returns

```python
goog_data['DAILY_RETURNS'] = goog_data[['Adj Close']].pct_change().fillna(0)
spy_data['DAILY_RETURNS'] = spy_data[['Adj Close']].pct_change().fillna(0)
```

```python
fig, axs = plt.subplots(1,2)
goog_data['DAILY_RETURNS'].plot(kind='hist',ax=axs[0],figsize=(15,5),bins=20,title="Google Daily Returns")
spy_data['DAILY_RETURNS'].plot(kind='hist',ax=axs[1],figsize=(15,5),bins=20,title="SnP Daily Returns")
```

### Comparing Histograms of Daily Returns

```python
goog_data['DAILY_RETURNS'].hist(bins=20,label="Google",figsize=(10,5))
spy_data['DAILY_RETURNS'].hist(bins=20,label="SnP")
plt.legend(loc='upper right')
```

## Scatter Plot - GOOG V/s SPY

Beta value is slope and alpha is intercept. 
* Alpha is very close to 0 which means on an average daily returns of GOOG performs similar to SPY.
* Beta shows reaction of GOOG with SPY. Which means on average GOOG returns 1.24% up for 1% return up in SPY.
* Note that Correlation is not equal to slope. Correlation is different than slope.
* Slope give direction while correlation is strength of relationship in any direction.

```python
beta_goog, alpha_goog = np.polyfit(spy_data['DAILY_RETURNS'],goog_data['DAILY_RETURNS'],1)
plt.figure(figsize=(10,5))
plt.scatter(spy_data['DAILY_RETURNS'],goog_data['DAILY_RETURNS'])
plt.plot(spy_data['DAILY_RETURNS'],beta_goog*spy_data['DAILY_RETURNS']+alpha_goog,'-',color='r')
plt.xlabel('SnP')
plt.ylabel('Google')
plt.title('Google Vs SnP Daily Returns')
print "Beta Google: " + str(beta_goog)
print "Alpha Google: " + str(alpha_goog)
print "Correlation Between SnP and Google: "+str(np.corrcoef(spy_data['DAILY_RETURNS'],goog_data['DAILY_RETURNS'])[0,1])
```
